```python
import shutil
import os

# Kaynak ve hedef klasör yolları
src = "/content/drive/MyDrive/Liver Vessel/nnunet/data/nnUNet_raw/Dataset001_hepaticvessel"
dst = "/content/drive/MyDrive/Liver Vessel/nnunet/data/nnUNet_raw/Dataset002_hepaticvessel"

# Hedef klasörü oluştur (varsa hata vermez)
os.makedirs(dst, exist_ok=True)

# İçeriği tek tek kopyala
for item in os.listdir(src):
    s = os.path.join(src, item)
    d = os.path.join(dst, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)  # klasörse recursive kopyala
    else:
        shutil.copy2(s, d)  # dosyaysa doğrudan kopyala

print("Kopyalama tamamlandı.")
```

```python
import os
import SimpleITK as sitk
from tqdm import tqdm

veela_train = "/content/drive/MyDrive/Liver Vessel/VEELA Train Set/VEELA Train Set"

dest_pathCT = "/content/drive/MyDrive/Liver Vessel/nnunet/data/nnUNet_raw/Dataset002_hepaticvessel/imagesTr"
dest_pathGT = "/content/drive/MyDrive/Liver Vessel/nnunet/data/nnUNet_raw/Dataset002_hepaticvessel/labelsTr"



for path in tqdm(os.listdir(veela_train)):
  if path.endswith("_norm.nii"):
    img = sitk.ReadImage(os.path.join(veela_train, path))
    new_id = path.replace("_norm.nii", "").replace("set", "")
    new_path = os.path.join(dest_pathCT, "hepaticvessel_" + str(int(new_id)+458).zfill(5) + "_0000.nii.gz")
    sitk.WriteImage(img, new_path)
  if path.endswith("_gt.nii"):
    img = sitk.ReadImage(os.path.join(veela_train, path))
    new_id = path.replace("_gt.nii", "").replace("set", "")
    new_path = os.path.join(dest_pathGT, "hepaticvessel_" + str(int(new_id)+458).zfill(5) + ".nii.gz")
    sitk.WriteImage(img, new_path)
```
